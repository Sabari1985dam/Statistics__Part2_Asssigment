1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
import numpy as np
import pandas as pd
import scipy.stats as st

sample_data = [50, 52, 48, 53, 51, 50, 54, 47, 61, 49]
population_mean = 50
population_std = 20
alpha = 0.05
n = len(sample_data)

sample_mean = np.mean(sample_data)

#calculate z-score
z_score = (sample_mean - population_mean)/(population_std/np.sqrt(n))

#calculate p-value
#Two Tail Test

p_value = 2*(1-st.norm.cdf(abs(z_score)))

if p_value < alpha:
  print("Reject the null Hypothesis")
else:
  print("Failed to reject the null hypothesis")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


pop_mean = 100
pop_std = 15
sample_size = 30

sample_data = np.random.normal(loc=105, scale=pop_std, size=sample_size)


sample_mean = np.mean(sample_data)


z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))


p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))


alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha/2)

if abs(z_score) > critical_value:
    print("Reject the null hypothesis (Hâ‚€): The sample mean significantly differs from the population mean.")
else:
    print("Fail to reject the null hypothesis (Hâ‚€): No significant difference.")


x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(8,6))
plt.plot(x, y, label="Standard Normal Distribution", color="blue")
plt.axvline(-critical_value, color="red", linestyle="dashed", label="Critical Value (-Z)")
plt.axvline(critical_value, color="red", linestyle="dashed", label="Critical Value (+Z)")
plt.axvline(z_score, color="green", linestyle="dashed", label=f"Observed Z-score ({z_score:.2f})")
plt.fill_between(x, y, where=(x < -critical_value) | (x > critical_value), color='red', alpha=0.3)
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.title("Decision Region in Two-Tailed Z-Test")
plt.legend()
plt.grid()
plt.show()
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(alpha=0.05, beta=0.20, mu_0=100, mu_1=105, sigma=15, n=30):

    se = sigma / np.sqrt(n)
    critical_value = mu_0 + stats.norm.ppf(1 - alpha) * se

    x = np.linspace(mu_0 - 4*sigma, mu_1 + 4*sigma, 1000)

    y_null = stats.norm.pdf(x, mu_0, se)
    y_alt = stats.norm.pdf(x, mu_1, se)


    beta_area = stats.norm.cdf(critical_value, mu_1, se)


    plt.figure(figsize=(10, 6))
    plt.plot(x, y_null, label="Null Distribution (Hâ‚€)", color="blue")
    plt.plot(x, y_alt, label="Alternative Distribution (Hâ‚)", color="red")


    plt.fill_between(x, y_null, where=(x > critical_value), color='blue', alpha=0.3, label="Type 1 Error (Alpha)")


    plt.fill_between(x, y_alt, where=(x < critical_value), color='red', alpha=0.3, label="Type 2 Error (Beta)")

    plt.axvline(critical_value, color="black", linestyle="dashed", label=f"Critical Value ({critical_value:.2f})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.title("Visualization of Type 1 and Type 2 Errors")
    plt.legend()
    plt.grid()
    plt.show()

    print(f"Critical Value: {critical_value:.2f}")
    print(f"Type 2 Error Probability (Beta): {beta_area:.4f}")


visualize_errors()
```

9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
def confidance_interval(sample_data, confidance = 0.95, population_std = None):
  size = len(sample_data)
  mean = np.mean(sample_data)
  std = np.std(sample_data)

  if population_std:
    z_score = stats.norm.ppf(1-(1-confidance)/2)
    margin_of_error = z_score * (population_std/np.sqrt(size))

  else:
    t_score = stats.t.ppf(1-(1-confidance)/2, df = size-1)
    margin_of_error = t_score * (std/np.sqrt(size))

  return mean-margin_of_error, mean+margin_of_error

sample_data = np.random.normal(loc = 20, scale = 25, size = 50)
low, up = confidance_interval(sample_data)
print(f"The confidance interval of the sample data is ({low:.2f},{up:.2f})")
```

16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
import scipy.stats as stats
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)


var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)


F_statistic = var1 / var2


df1 = len(sample1) - 1
df2 = len(sample2) - 1
p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (Hâ‚€): Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis (Hâ‚€): No significant difference in variances.")
```

22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

sample1 = np.random.normal(loc=50, scale=5, size=30)
sample2 = np.random.normal(loc=55, scale=5, size=30)
sample3 = np.random.normal(loc=60, scale=5, size=30)

f_statistic, p_value = stats.f_oneway(sample1, sample2, sample3)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (Hâ‚€): At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis (Hâ‚€): No significant difference between group means.")

plt.figure(figsize = (10,6))
sns.boxplot(data = [sample1, sample2, sample3])
plt.title("Group Means")
plt.xlabel("Samples/Groups")
plt.ylabel("Values")
plt.show()
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
n = 100
sample_std = 20
sample_mean = 70
sample = np.random.normal(loc = sample_mean, scale = sample_std, size = n)

sample_var = np.var(sample, ddof=1)
pop_var = 100

chi2_stat = (n-1)*sample_var/pop_var

p_value = 2*min(stats.chi2.cdf(chi2_stat, df = n-1), 1-stats.chi2.cdf(chi2_stat, df = n-1))

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (Hâ‚€).")
else:
    print("Fail to reject the null hypothesis (Hâ‚€).")
```

25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
import scipy.stats as stats


successes_group1 = 150
total_group1 = 1000

successes_group2 = 180
total_group2 = 1200


p1 = successes_group1 / total_group1
p2 = successes_group2 / total_group2


p_combined = (successes_group1 + successes_group2) / (total_group1 + total_group2)
se = np.sqrt(p_combined * (1 - p_combined) * (1/total_group1 + 1/total_group2))


z_score = (p1 - p2) / se


p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (Hâ‚€).")
else:
    print("Fail to reject the null hypothesis (Hâ‚€).")
```

26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=50, scale=15, size=100)

var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

F_statistic = var1 / var2
df1 = len(data1) - 1
df2 = len(data2) - 1
p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (Hâ‚€): The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis (Hâ‚€): No significant difference in variances.")


plt.figure(figsize=(8,6))
plt.hist(data1, bins=15, alpha=0.6, label="Dataset 1 (Std=10)", color="blue", density=True)
plt.hist(data2, bins=15, alpha=0.6, label="Dataset 2 (Std=15)", color="red", density=True)
plt.axvline(np.mean(data1), color="blue", linestyle="dashed", linewidth=2, label=f"Mean Dataset 1 ({np.mean(data1):.2f})")
plt.axvline(np.mean(data2), color="red", linestyle="dashed", linewidth=2, label=f"Mean Dataset 2 ({np.mean(data2):.2f})")
plt.title("Comparison of Two Normally Distributed Datasets")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
